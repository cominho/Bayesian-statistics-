```python
import numpy as np
from scipy.stats import dirichlet
import matplotlib.pyplot as plt
```

```python
# Why Dirichlet prior promotes sparsity

# Generate samples from Dirichlet distribution
def generate_dirichlet_samples(alpha, num_samples=1000):
    """Generate samples from Dirichlet distribution with a given alpha."""
    dimension = len(alpha)
    samples = dirichlet.rvs(alpha, size=num_samples)
    return samples


# Plot Dirichlet samples on a simplex
def plot_simplex(samples, alpha, ax):
    """Plot Dirichlet samples on a 2D simplex."""
    ax.scatter(samples[:, 0], samples[:, 1], alpha=0.5, s=5, label=f"α = {alpha[0]}")
    ax.set_xlim(0, 1)
    ax.set_ylim(0, 1)
    ax.set_title(f"α = {alpha[0]}")
    ax.set_xlabel("Dimension 1")
    ax.set_ylabel("Dimension 2")
    ax.set_aspect('equal')


def plot_dirichlet_simplex():
    """Generate and plot Dirichlet samples for different alpha values."""
    alphas = [[3, 3, 3], [1, 1, 1], [1/3, 1/3, 1/3]]
    fig, axes = plt.subplots(1, 3, figsize=(15, 5), sharex=True, sharey=True)

    for i, alpha in enumerate(alphas):
        samples = generate_dirichlet_samples(alpha, num_samples=500)
        # Project the 3D simplex to 2D
        samples_2d = np.array([s[:2] for s in samples])
        plot_simplex(samples_2d, alpha, axes[i])

    plt.tight_layout()
    plt.show()

plot_dirichlet_simplex()
```
